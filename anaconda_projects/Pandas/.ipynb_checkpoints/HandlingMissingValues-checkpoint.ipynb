## Handling Missing Values

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
data.head()
```

## 1. Dropping missing data using dropna()

```python
data.shape
```

```python
# if how equals to all it means ki agar saare columns meh empty data hai toh drop kardo
data.dropna(how='all').shape
```

```python
# It drop all columns which atleast have one empty value
data.dropna(axis=1).shape
```

```python
data.dropna(subset=['Cabin', 'Embarked']).shape
```

## 2. Filling missing values using fillna()

```python
data.fillna(0)
```

```python
# data['Cabin'].fillna("Not Specified", inplace=True) this is not working so new method below
data.fillna({"Cabin": "Not Specified"}, inplace=True)
```

```python
data
```

```python
data['Age'].fillna(30)
```

```python
# data['Age'].fillna(method='bfill')
data.fillna({"Cabin": "Not Specified"}, inplace=True)
```
