# Scatter Plots

The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

Although, scatter plot can be plotted for both numerical as well as categorical data, it's mostly used to depict relationship between numerical data.

There are 2 ways to draw a scatterplot in seaborn

1. relplot [Figure Level Function]

2. scatterplot [Axes Level Function]

As we will see, these visualizations can be quite helpful because they use simple and easily-understood representations of data even on complex data by mapping up to three additional variables using the semantics of hue, size, and style.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

## Plot a Scatter between total_bill and tip

```python
# Plot here
sns.relplot(x='total_bill', y='tip', kind='scatter', data=data)
```

### The hue parameter

```python
# Plot here
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

 ### The style parameter

```python
# Plot here
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', data=data)
```

### The size parameter

```python
# Plot here
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', size='size', sizes=(15, 200), data=data)
```

### Using the scatterplot function

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', style='sex', data=data)
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(
    x='day',
    y='tip',
    kind='strip',
    data=tips,
    hue='day',         # Explicitly map color to the 'day' column
    palette='Set2',
    legend=False,       # Hide the unnecessary legend
    jitter=0
)
# jitter  seperate these point so that they are not on top of each other this noise seperating them
```

```python
sns.catplot(x='day', y='tip', kind='swarm', data=tips, hue='day', palette='Set2')
```

```python
sns.catplot(x='day', y='tip', kind='swarm', hue='sex', data=tips)
```

```python
sns.swarmplot(x='day', y='tip', data=tips)
```

## Boxplots
A boxplot is a standardized way of displaying the distribution of data based on a five number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum"). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

```python
# univariate

sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='box', data=tips)
```

## Violin Plots
violin plots are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot. <br/>
The unquestionable advantage of the volin plot over the box plot is that aside from showing the above mentioned statistics it also shows the entire distribution of the data

Volin plot atre often used to compare the distribution of a given variable across some categories

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips, hue='day')
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips, hue='sex', split=True)
```

## Barplot and Countplot

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='smoker', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
```

```python
# for median estimator is median
sns.catplot(x='smoker', y='total_bill', hue='sex', estimator=np.median, kind='bar', data=tips)
```

```python
# for variace estimator is var
sns.catplot(x='smoker', y='total_bill', hue='sex', estimator=np.var, kind='bar', data=tips)
```

```python
sns.catplot(x='sex',  hue='smoker',kind='count', data=tips)
```

## Heatmaps
A heat map is a graphical represntation of data where the individual values contained in a matrix are represented as colors

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
#If you want your pivot table to continue including all possible categories, regardless of whether they have data, 
#add observed=False to your code.
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum', observed=False)
```

```python
# to hide bar show color with values cbar=false
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
```

```python
# for colorblind
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

## Clustermap

```python
 sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot=True)
```

```python
sns.clustermap(x, metric='correlation', z_score=0, annot=True, row_cluster=False)
```

## JointPlot
Jointplot creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes. <br/>
We can use Jointplots in the following 4 manners: <br/>
1. ScatterPlots <br/>
2. Hexbin Plots <br/>
3. KdePlots <br/>
4. RegPlot

```python
tips = sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

## Pairplots
Pairplots are used to plot multiple pairwise bivariate distributions in a dataset. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. By default, it also draws the univariate distribution of each variable on the diagonal Axes.

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

## Displots
Distplots are used to do univariate analysis on a continous/numerical data. <br/>
Distplot combines 3 things in one graph: <br/>
1. Histogram <br/>
2. Kernel Density Enstimation (PDF) <br/>
3. Rugplot

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

### Plotting a Distplot

```python
sns.distplot(titanic['age'])
```

### Plotting only the Histogram

```python
sns.histplot(titanic['age'], kde=False)
```

```python
sns.histplot(titanic['age'], bins=100, kde=False)
```

### Plotting onlt the KDE

```python
sns.displot(titanic['age'], kind='kde')
```

### Plotting onlt the Rugplot

```python
# sns.displot(titanic['age'], hist=False, kde=False, rug=True)
# new method old is now working
sns.rugplot(data=titanic, x='age')
```

### How to use the Distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
