```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['Embarked'])
```

## b. Piechart

```python
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

## b. Distplot

```python
sns.distplot(df['Age'])
```

## c. Boxplot



```python
# it is use to hive 5 number summary
sns.boxplot(df['Age'])
```
