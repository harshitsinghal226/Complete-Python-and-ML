```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
df = df.iloc[:,1:]
```

```python
# Steps

# 0. Preprocess + EDA + Feature Selection
# 1. Extract input and output cols
# 2. Scale the values
# 3. Train test split
# 4. Train the model
# 5. Evaluate the model/model selection
# 6. Deploy the model
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['cgpa'], df['iq'])
```

```python
# c=df['placement'] is to color on bases of placement 
plt.scatter(df['cgpa'], df['iq'], c=df['placement'])
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
# For Train test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.1)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
# For scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
# for training model
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# model training
clf.fit(X_train, Y_train)
```

```python
# model evaluation
y_pred = clf.predict(X_test)
```

```python
Y_test
```

```python
# for checking accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, y_pred)
```

```python
# agar check karna hai ki machine learning kya pattern use kara hai toh usko visualize karne ke liye 
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_region(X_train.values, Y_train.values, clf=clf, legend=2)
```
